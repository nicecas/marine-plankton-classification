```python
import glob
import torch
from torchvision.models.feature_extraction import get_graph_node_names

# Get .pt files containing "best" in the file name
model_files = glob.glob('../2-TrainingDNN/models/*best*.pt')

# Load each model and print its graph node names
for model_file in model_files:
    model = torch.load(model_file, weights_only=False)  # Load the model file
    nodes, _ = get_graph_node_names(model)
    print(f"Nodes for {model_file}:")
    
    # If there are more than 5 nodes, print '...' followed by the last 5 nodes
    if len(nodes) > 5:
        print("... " + ",".join(nodes[-5:]))
    else:
        print(",".join(nodes))
        
    print("\n" + "-" * 40 + "\n")
```
